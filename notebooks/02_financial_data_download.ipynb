```python
from src.loading_tickers import load_tickers_from_website
tickers = load_tickers_from_website(save_to_file = True, file_name = 'WIG_ticker_list')['Nazwa']
```

```python
from pathlib import Path
cur_dir = Path.cwd()
DOWNLOAD_DIR = cur_dir.parent / 'data/financial_data'
```

```python
from src.scraping import GetData
import time 
import random 



errors = {}
for i, cname in enumerate(tickers): 
    print(f'{i+1}/{len(tickers)}: Pobieranie danych dla: {cname}')
    try:
        cid, isbank, df = GetData(cname)
        df.to_csv(DOWNLOAD_DIR / f'{cname}.csv')
    except Exception as e:
        print(f'Błąd przy pobieraniu {cname}: {e}')
        errors[cname] = e
    print('')
    time.sleep(random.uniform(3, 7))

if errors:
    print(f'Liczba błędów: {len(errors)}')
    print(errors)
```
