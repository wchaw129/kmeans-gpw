```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from src.processing import set_plot_style, set_plot_font, save_chart

sector_map = {
    'ALIOR': 'Finanse', 'MBANK': 'Finanse', 'MILLENNIUM': 'Finanse', 
    'PEKAO': 'Finanse', 'PKOBP': 'Finanse', 'SANPL': 'Finanse',
    'PZU': 'Finanse', 'KRUK': 'Finanse', 'XTB': 'Finanse',
    'ALLEGRO': 'Handel i Usługi', 'CCC': 'Handel i Usługi', 'DINOPL': 'Handel i Usługi', 
    'LPP': 'Handel i Usługi', 'PEPCO': 'Handel i Usługi', 'ZABKA': 'Handel i Usługi', 
    'BENEFIT': 'Handel i Usługi', 'RAINBOW': 'Handel i Usługi',
    'BUDIMEX': 'Przemysł i Energetyka', 'KETY': 'Przemysł i Energetyka', 
    'ENEA': 'Przemysł i Energetyka', 'PGE': 'Przemysł i Energetyka', 
    'TAURONPE': 'Przemysł i Energetyka', 'PKNORLEN': 'Przemysł i Energetyka', 
    'JSW': 'Przemysł i Energetyka', 'KGHM': 'Przemysł i Energetyka',
    'ASSECOPOL': 'Technologia', 'CDPROJEKT': 'Technologia', 
    'CYFRPLSAT': 'Technologia', 'ORANGEPL': 'Technologia', 'SYNEKTIK': 'Technologia'
}

set_plot_style()
set_plot_font()
sector_data = pd.Series(sector_map)
plt.figure(figsize = (8, 4))
sns.countplot(x = sector_data, width = 0.55, palette="binary")

plt.yticks(np.arange(11))
plt.xlabel('')
plt.ylabel('liczba spółek w sektorze')
save_chart('chart4')
```
