```python
from src.loading_tickers import load_tickers_from_website

url_wig30 = r'https://www.bankier.pl/inwestowanie/profile/quote.html?symbol=WIG30'
TO_DIR = r'../data/tickers.csv'
cnames_table = load_tickers_from_website(url_wig30)[['Nazwa', 'Ticker']]
cnames_table.to_csv(TO_DIR)
cnames_table.head()
```

```python
from src.indicators import count_indicators
import pandas as pd
company_names = cnames_table['Nazwa']
tickers = cnames_table['Ticker']
df = pd.DataFrame()

i=1
errors=0
for cname, ticker in zip(company_names, tickers):
    print(f'Pobieranie {i}/30')
    try:
        info = pd.Series(count_indicators(cname, ticker, 3, 5, 2024))
    except Exception as e:
        print(f'Błąd dla spółki: {cname} \n {e} \n')
        errors+=1
    df[cname] = info
    i+=1

print(f'Liczba błędów: {errors}')
```

```python
data = df.transpose()

TO_DIR = r'../data/dirty/dirty_df.csv'
data.to_csv(TO_DIR)
```
