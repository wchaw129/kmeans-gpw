Przykłady do rozdziału teoretycznego

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
src = r"../data/example_data.csv"
data = pd.read_csv(src)[['X', 'Y']]
print(data.head())
```

```python
data.plot(x='X', y='Y', kind='scatter')
plt.show()
```

```python
from src.processing import set_plot_style, set_plot_font, save_chart
inertia_values = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    inertia_values.append(kmeans.inertia_)

set_plot_font()
plt.plot(range(1, 11), inertia_values, marker='o', color = 'grey', linestyle = '--')
plt.xlabel('Liczba klastrów (k)')
plt.ylabel('Inercja')
plt.xticks(range(1, 11))
save_chart('example_elbow')
```

```python
from sklearn.metrics import silhouette_score
shilhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data)
    data['labels'] = kmeans.labels_
    a = silhouette_score(data[['X', 'Y']], data['labels'])
    shilhouette_scores.append(a)
plt.plot(range(2, 11), shilhouette_scores, marker='o')
plt.xlabel('k')
plt.ylabel(r'$\bar{S}$(k)')
plt.xticks(range(2, 11))
plt.show()
```
